```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, mean_squared_error, r2_score
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

import numpy as np
```

```python
# Lectura y pre-procesado

df = pd.read_csv("medical_insurance.csv")

df.head()
df.info()
df.describe(include="all")

```

```python
# Comprobar nulos
df.isna().sum().sort_values(ascending=False).head(20)

# Rellenamos nulos con mediana para podeer hacer uso de todas las variables y las categoricas con la moda por lo mismo
num_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
cat_cols = df.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])


df.info()
```

```python
# DASHBOARD EDA
vars_clinicas = ["bmi", "systolic_bp", "diastolic_bp", "ldl", "hba1c", 
                 "risk_score", "annual_medical_cost"]
corr = df[vars_clinicas].corr()

# Crear figura y ejes
fig, axes = plt.subplots(2, 3, figsize=(18, 8))

df["age_group"] = pd.cut(df["age"], bins=[0,30,45,60,80,120], labels=["<30","30-45","45-60","60-80","80+"]) ##SEGMENTAR POR LOS GRUPOS DE EDAD

# Histograma coste médico anual
sns.histplot(df["annual_medical_cost"], bins=50, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Distribución del coste médico anual")
axes[0, 0].set_xlabel("annual_medical_cost")
axes[0, 0].set_ylabel("Frecuencia")

# Coste medio por grupo de edad y sexo
sns.barplot(data=df, x="age_group", y="annual_medical_cost",
            hue="sex", estimator=np.mean, ax=axes[0, 1])
axes[0, 1].set_title("Coste medio por grupo de edad y sexo")
axes[0, 1].set_xlabel("Grupo de edad")
axes[0, 1].set_ylabel("Coste medio")
axes[0, 1].legend(title="Sexo", fontsize=8)

# Coste medio según hábito de fumar
sns.barplot(data=df, x="smoker", y="annual_medical_cost",
            estimator=np.mean, ax=axes[0, 2])
axes[0, 2].set_title("Coste medio según hábito de fumar")
axes[0, 2].set_xlabel("Smoker")
axes[0, 2].set_ylabel("Coste medio anual")

# Días hospitalizado vs total pagado en reclamaciones
sample_df = df.sample(min(5000, len(df)), random_state=0)  # por si hay <5000 filas
sns.scatterplot(data=sample_df,
                x="days_hospitalized_last_3yrs",
                y="total_claims_paid",
                alpha=0.3,
                ax=axes[1, 0])
axes[1, 0].set_title("Días hospitalizado vs total pagado en reclamaciones")
axes[1, 0].set_xlabel("Días hospitalizado (últimos 3 años)")
axes[1, 0].set_ylabel("Total pagado en reclamaciones")

# Matriz de correlación variables clínicas y coste
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", ax=axes[1, 1])
axes[1, 1].set_title("Matriz de correlación (clínicas y coste)")


plt.tight_layout()
plt.show()
```

```python
#Clustering

# Selección de variables para clustering  NO ESCOGEMOS TODAS PORQUE SOLO QUEREMOS VER LAS QUE PARECEN
cluster_vars = [
    "age", "bmi", "chronic_count", "visits_last_year",
    "hospitalizations_last_3yrs", "annual_medical_cost", "total_claims_paid"
]

X_cluster = df[cluster_vars].copy() ## para evitar cargarme lo anterior

# Escalado
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

kmeans = KMeans(n_clusters=4, random_state=0)
df["cluster"] = kmeans.fit_predict(X_scaled)

pca = PCA(n_components=2, random_state=0)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(6,5))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=df["cluster"], palette="tab10", alpha=0.5)
plt.title("Clusters de pacientes (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()

```

```python
# Perfiles del Cluster
df.groupby("cluster")[cluster_vars].mean().round(2)
```

```python
#### MODELO DE ML

target = "is_high_risk" ### VARIABLE DE RIESGO ALTO PARA PODER PREDECIR EL RIESGO DE LOS PACIENTES.

features = [
    "age", "sex", "bmi", "smoker",
    "chronic_count",
    "visits_last_year",
    "plan_type"
]

X = df[features].copy()
y = df[target].astype(int) 

# Definir tipos de columnas
num_features = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
cat_features = X.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

numeric_transformer = Pipeline(steps=[
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_features),
        ("cat", categorical_transformer, cat_features)
    ]
)

clf = LogisticRegression(max_iter=1000)

model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", clf)
])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0,
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Coeficientes 
ohe = model.named_steps["preprocessor"].named_transformers_["cat"]["onehot"]
ohe_features = ohe.get_feature_names_out(cat_features) ## features del encoder (VARIABLES CATEGORICAS!)

feature_names = np.concatenate([num_features, ohe_features])

coefs = model.named_steps["classifier"].coef_[0]

coef_df = pd.DataFrame({
    "feature": feature_names,
    "coef": coefs
}).sort_values("coef", key=abs, ascending=False)

coef_df.head(20)
```
